```python
import turtle
import time
import random


# Set up the screen
screen = turtle.Screen()
screen.setup(width=600, height=600)
screen.bgcolor("black")

# Create lane turtle for drawing static elements
lane_turtle = turtle.Turtle()
lane_turtle.hideturtle()
lane_turtle.speed(0)
lane_turtle.color("white")
lane_turtle.pensize(5)

# Function to draw a line between two points, adjusted for center at (0, 220)
def draw_line(x_start, y_start, x_end, y_end):
    lane_turtle.penup()
    lane_turtle.goto(x_start, y_start - 220)
    lane_turtle.pendown()
    lane_turtle.goto(x_end, y_end - 220)

# Draw the path layout
draw_line(0, 0, 0, 40)
draw_line(-200, 0, -200, 440)
draw_line(-200, 40, 200, 40)
draw_line(-200, 240, 200, 240)
draw_line(-200, 440, 200, 440)
draw_line(70, 440, 70, 340)
draw_line(0, 440, 0, 240)
draw_line(0, 340, -70, 340)
draw_line(200, 0, 200, 440)

# Create initial rubbish circles at specified locations
rubbish_positions = [(0, 40,random.randint(0, 1)), (0, 240,random.randint(0, 1)), (-60, 100,random.randint(0, 1)), (70, 150,random.randint(0, 1)), (80, 240,random.randint(0, 1)), (80, 40,random.randint(0, 1))]
rubbish_circles = []

for pos in rubbish_positions:
    r = turtle.Turtle()
    r.shape("circle")
    if (pos[2]==0):
        r.color("green")
        rubbish_circles.append([r,"green"])
    else:
        r.color("red")
        rubbish_circles.append([r,"red"])
    r.penup()
    r.goto(pos[0], pos[1] - 220)  # Adjust for center at (0, 220)

# Function to create green and red zones
def draw_square(x_start, y_start, width, height, color):
    square_turtle = turtle.Turtle()
    square_turtle.hideturtle()
    square_turtle.speed(0)
    square_turtle.color(color)
    square_turtle.penup()
    square_turtle.goto(x_start, y_start - 220)
    square_turtle.pendown()
    square_turtle.begin_fill()
    for _ in range(2):
        square_turtle.forward(width)
        square_turtle.right(90)
        square_turtle.forward(height)
        square_turtle.right(90)
    square_turtle.end_fill()

# Draw the red and green zones
draw_square(40, 370, 60, 60, "red")
draw_square(-100, 370, 60, 60, "green")

# Create the yellow turtle car
car = turtle.Turtle()
car.shape("turtle")
car.color("yellow")
car.penup()
car.speed(1)


car.goto(0, - 220)
car.setheading(90)
car.forward(40)
rub = rubbish_circles[0]
rubbish = rub[0]
rubbish.hideturtle()
car.left(90)
def onetog():
    car.forward(200)
    car.right(90)
    car.forward(400)
    car.right(90)
    car.forward(200)
    car.right(90)
    car.forward(100)
    car.right(90)
    car.forward(70)
    car.backward(70)
    car.left(90)
    car.forward(100)
    
def onetor():
    car.forward(200)
    car.right(90)
    car.forward(400)
    car.right(90)
    car.forward(270)
    car.right(90)
    car.forward(100)
    car.backward(100)
    car.right(90)
    car.forward(70)
    car.left(90)
    car.forward(200)

if rubbish_circles[0][1]=="green":
    onetog()
else:
    onetor()

rub = rubbish_circles[1]
rubbish = rubbish_circles[1][0]
rubbish.hideturtle()
def twotog():
    car.backward(100)
    car.right(90)
    car.forward(70)
    rubbish.goto(car.position())
    rubbish.showturtle()
    car.backward(70)
    car.left(90)
    car.forward(100)

    
def twotor():
    car.backward(200)
    car.left(90)
    car.forward(70)
    car.right(90)
    car.forward(100)
    rubbish.goto(car.position())
    rubbish.showturtle()
    car.backward(100)
    car.right(90)
    car.forward(70)
    car.left(90)
    car.forward(200)

rub=rubbish_circles[1]
rubbish = rubbish_circles[1][0]
if rub[1]=='green':
    twotog()
else:
    twotor()

car.left(90)
car.forward(80)
rub = rubbish_circles[4]
rubbish = rubbish_circles[4][0]
rubbish.hideturtle()
car.backward(80)
car.right(90)

if rub[1]=='green':
    twotog()
else:
    twotor()

car.left(90)
car.forward(70)
car.right(120)
car.left(60)
car.right(30)
car.forward(90)
rub = rubbish_circles[3]
rubbish = rubbish_circles[3][0]
rubbish.hideturtle()
car.backward(90)
car.right(90)
car.forward(70)
car.left(90)

if rub[1]=='green':
    twotog()
else:
    twotor()

car.left(90)
car.forward(200)
car.right(90)
car.forward(200)
car.right(90)
car.forward(120)
rub = rubbish_circles[5]
rubbish = rubbish_circles[5][0]
rubbish.hideturtle()
car.forward(80)

if rub[1]=="green":
    onetog()
else:
    onetor()

car.right(90)
car.forward(200)
car.left(90)
car.forward(200)
car.left(90)
car.forward(140)
car.left(120)
car.right(60)
car.left(30)
car.forward(60)
rub = rubbish_circles[2]
rubbish = rub[0]
rubbish.hideturtle()
car.backward(60)
car.left(90)
car.backward(60)

if rub[1]=="green":
    onetog()
else:
    onetor()

car.left(90)
car.forward(200)
car.right(90)
car.forward(200)
car.right(90)
car.forward(200)
car.right(90)
car.backward(40)

turtle.done()
```
